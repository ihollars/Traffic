### Imports

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
colors = sns.color_palette("deep")
```

## 4-way stop

```python
naive = pd.read_csv('4way_naive destinations-table.csv')
naive.columns = ['run_number','spawn_prob', 'complete' ]
naive.drop('run_number', axis=1, inplace=True)

# plot a bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="spawn_prob", y="complete", data=naive, estimator=np.mean, errorbar='sd', capsize=.2, color=colors[0])
ax.set_xlabel('Spawn Probability', weight='bold', fontsize=14)
ax.set_ylabel('Number of Completed Destinations', weight='bold', fontsize=14)
ax.set_title('4-way Stop Efficiency', weight='bold',fontsize=18)

plt.savefig('4-way Stop Efficiency.png')
```
